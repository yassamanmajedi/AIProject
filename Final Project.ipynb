_________________________________
<h1>Standardizing the Data</h1>

```python
scaler=StandardScaler()
scaler.fit(x)
```

```python
scaled_data=scaler.transform(x)
scaled_data
```

__________________
<h1>Apply PCA</h1>

```python
pca=PCA(n_components=0.85)
# pca.fit(scaled_data)
# x_pca = pca.transform(scaled_data)
x_pca=pca.fit_transform(scaled_data)
```

__________________________________
<h1>Explained Variance</h1>

The explained variance tells us how much information (variance) can be attributed to each of the principal component

```python
pca.explained_variance_ratio_
```

```python
pca=PCA().fit(scaled_data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of principal components')
plt.ylabel('cumulative explained variance')
```

___________________
<h1>Apply Support Vector Machine to the Transformed Data</h1>

```python
from sklearn.model_selection import train_test_split
from sklearn import svm

X_pca_train, X_pca_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=42)

clf=svm.SVC(kernel='linear')
clf.fit(X_pca_train, y_train)
pred = clf.predict(X_pca_test)

clf.score(X_pca_test, y_test)
```

___________________
<h1>Apply KNeighbors Classifier to the Transformed Data</h1>

```python
from sklearn.neighbors import KNeighborsClassifier

clf= KNeighborsClassifier(n_neighbors=10,)
clf.fit(X_pca_train, y_train)
pred = clf.predict(X_pca_test)

clf.score(X_pca_test, y_test)
```

___________________
<h1>Apply Decision Tree to the Transformed Data</h1>

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(X_pca_train, y_train)
pred = clf.predict(X_pca_test)

clf.score(X_pca_test, y_test)
```
